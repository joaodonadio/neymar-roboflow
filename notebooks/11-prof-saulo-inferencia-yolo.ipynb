## 2. Carregando o Modelo Específico
Agora instanciamos o modelo passando o nosso próprio arquivo com os "pesos" (`modelo_aquario.pt`).

```python
import os
from ultralytics import YOLO

# 1. Pega os caminhos corretos (subindo um nível a partir da pasta /notebooks)
pasta_do_notebook = os.getcwd()
raiz_do_projeto = os.path.dirname(pasta_do_notebook)

# 2. Aponta para o nome REAL do seu arquivo de modelo na raiz do projeto
nome_modelo = "modelo-neymar-best.pt"
caminho_completo_modelo = os.path.join(raiz_do_projeto, nome_modelo)

print(f"Carregando modelo: {caminho_completo_modelo}...")

# 3. Carrega o seu modelo personalizado do Neymar
model = YOLO(caminho_completo_modelo)

# 4. Verifica quais classes esse modelo conhece (deve aparecer algo relacionado ao Neymar)
print("\nClasses que este modelo conhece:", model.names)
```

## 3. Realizando a Inferência
Opcional: você pode baixar um arquivo mp4 e colocar na mesma pasta, ou testar em uma imagem da internet se o vídeo local não estiver disponível.

```python
import os
import cv2
from ultralytics import YOLO

# 1. Escolha a sua origem (Dica: Links de vídeos funcionam melhor para testar pós-clone)
# Você pode usar o link do youtube que testou antes: "https://www.youtube.com/watch?v=JS6v6vkrNEU"
source = "https://upload.wikimedia.org/wikipedia/commons/e/e4/A_Shoal_of_Fish.jpg"

# Se preferir testar um arquivo local, descomente a linha abaixo e comente a de cima:
# source = os.path.join(os.getcwd(), 'aquario.mp4')

# 2. Validação extra: Se for arquivo local, testa se o OpenCV realmente consegue abrir
if isinstance(source, str) and os.path.exists(source) and source.endswith(('.mp4', '.avi', '.mov')):
    cap = cv2.VideoCapture(source)
    if not cap.isOpened():
        print("⚠️ O arquivo de vídeo existe, mas está corrompido ou sem codecs. Mudando para link online...")
        source = "https://upload.wikimedia.org/wikipedia/commons/e/e4/A_Shoal_of_Fish.jpg"
    cap.release()

print(f"🎬 Origem definida para o YOLO: {source}")

# 3. Rodar a detecção com stream=True para não estourar a RAM
# O 'model' deve ser o seu modelo do Neymar carregado na célula anterior
resultados = model.predict(source=source, show=False, save=True, conf=0.5, stream=True)

print("🚀 Processando... (Se for um vídeo longo, clique no '⬛' do notebook para parar quando quiser)")

# 4. Consome os frames um por um com segurança
try:
    for frame in resultados:
        pass
    print("🎉 Processamento concluído com sucesso!")
except ValueError as e:
    print(f"❌ Erro ao processar a mídia: {e}")
    print("💡 Sugestão: Altere a variável 'source' lá no topo para um link da internet válido.")
```
