Este notebook gerencia o pipeline completo, desde a verificação do hardware (GTX 1660 Ti), passando pelo treinamento do dataset do Roboflow, até a inferência em vídeos do YouTube corrigindo bugs de streaming.

```python
# Passo 1: Verificar se o ambiente está usando a GPU (GTX 1660 Ti) corretamente
import torch
import ultralytics

print(f"Versão do PyTorch: {torch.__version__}")
print(f"Versão do Ultralytics: {ultralytics.__version__}")
print("CUDA está disponível?", torch.cuda.is_available())

if torch.cuda.is_available():
    print(f"Placa de Vídeo Detectada: {torch.cuda.get_device_name(0)}")
else:
    print("⚠️ ALERTA: CUDA não detectado. O código vai rodar na CPU (lento).")
```

## 🚀 1. Treinamento do Modelo (Treinar o "Rei")
Execute esta célula se precisar treinar o modelo novamente ou ajustar os parâmetros. O resultado final (best.pt) será salvo na pasta runs/detect/train/weights/.

```python
import os
from ultralytics import YOLO

# Pega a pasta onde o notebook está (ex: /home/joaov/github/neymar-roboflow/notebooks)
pasta_do_notebook = os.getcwd()

# Sobe um nível para ir para a raiz do projeto (/home/joaov/github/neymar-roboflow)
raiz_do_projeto = os.path.dirname(pasta_do_notebook)

# Constrói o caminho correto apontando para a raiz
caminho_data = os.path.join(
    raiz_do_projeto, 
    "dataset-ney", 
    "neymar.v1i.yolov8", 
    "data.yaml"
)

# Carrega o modelo base
modelo_base = YOLO("yolov8n.pt") 

# Inicia o treino
modelo_base.train(
    data=caminho_data,
    epochs=50,          
    imgsz=640,          
    batch=16,           
    device=0,           
    workers=4,          
    save=True
)
print("🎉 Treinamento concluído com sucesso!")
```
